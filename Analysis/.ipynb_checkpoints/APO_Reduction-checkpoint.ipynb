```python
file = science_dir

for name in glob.glob(file + '*.fits'):

    hdul = fits.open(name)
    
    data = hdul[0].data
    header = hdul[0].header
    
    std = np.std(hdul[0].data)
    mean = np.mean(hdul[0].data)
    
    print(f"std: {std}")
    print(f"mean: {mean}")
    print(f"exptime: {header['exptime']}")
    
    plt.imshow(hdul[0].data,vmin=(mean-std),vmax=(mean+std))
    plt.title(name)
    plt.colorbar()
    plt.show()
    plt.close()
```
