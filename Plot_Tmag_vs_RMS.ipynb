```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
```

```python
data = ascii.read('Data/tess_rms.dat')
tmags = data['Tmag']
TESS_rms = data['TESS_rms']
```

```python
LCO_rms = [0.0014366,0.0015638,0.0029571,
           0.0008963,0.000318,0.000318,0.0005331,0.000999,0.0003532,
           0.0010103,0.0008675,0.0007380,0.0029077,0.0016818,
           0.0012413,0.0007272,0.0010043,0.0022966,0.0008824,0.0005496]
LCO_err = [0.0003212,0.0002681,0.0005588,
           0.0001911,0.0000678,0.0000678,0.0002176,0.0001824,0.000079,
           0.0001559,0.0001308,0.0001167,0.0004197,0.0002884,
           0.0002069,0.0001818,0.0001722,0.0003631,0.0001667,0.0000943]
Tmag = [13.60,13.04,14.88,
        11.90,9.39,9.39,9.44,13.40,12.27,13.24,13.3,12.38,
        14.79,13.59,13.51,13.51,13.497,15.00,13.46,13.05]
```

```python
plt.figure(figsize=(6,5))
plt.scatter(tmags,TESS_rms,label='TESS',alpha = 0.4,color='k',s=10)
plt.errorbar(Tmag[-2:],LCO_rms[-2:],yerr=LCO_err[-2:],label='LCO',alpha = 1.0,color='royalblue',fmt='o')
plt.legend()
plt.title('LCO can outperform TESS for faint stars')
plt.ylim(0.0001,0.01)
plt.xlim(7,15)
plt.yscale('log')
plt.xlabel('TESS Magnitude')
plt.ylabel(r'Light Curve RMS (10-minute timescale)')
plt.tight_layout()
plt.savefig('Figs/RMS_comparison.pdf')
```
