```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
focus = [0,0,0,1,1,1,2,2,2,3,3,3]
HWHM_radius = [2.3,2.3,2.1,4,4.1,4,7.8,7.7,7.7,10.1,9.9,9.4]
Star_radius = [9,10,9,13,13,13,18,18,18,24,24,24]

plt.figure(figsize=(10,8))
plt.scatter(focus,HWHM_radius,alpha=0.5,label='HWHM',color='black')
plt.scatter(focus,Star_radius,alpha=0.5,label='Object Radius')
plt.xlabel('Focus offset (mm)')
plt.ylabel('Radius (pixels)')
plt.title('Defocusing with LCO')
plt.legend()
plt.savefig('LCO_Defocusing')
```
